```python
# === Four-helix bundle quantification (DSSP) for *.pdb in current directory ===
# deps: mkdssp (system), pip install biopython pandas numpy matplotlib

import os, glob, math, shutil, re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from Bio.PDB import PDBParser
from Bio.PDB.DSSP import DSSP

# -----------------------------
# Configuration
# -----------------------------
INPUT_GLOB          = "./*.pdb"      # scan current directory (e.g., 1.pdb, 2.pdb, ...)
CHAIN_ID            = None           # None = use all protein chains; or set 'A' if you know
HELIX_CODES         = set(["H","G","I"])  # DSSP helix types: alpha, 3-10, pi
MIN_HELIX_LEN       = 8              # residues; segment length to count as helix
REQ_N_HELICES       = 4              # must have ≥ this many helices to be a 4HB
ANGLE_RANGE_DEG     = (40.0, 80.0)   # acceptable mean pairwise axis angle (degrees)
ANGLE_SD_MAX        = 30.0           # max SD of pairwise angles (degrees)
BUNDLE_RADIUS_MAX_A = 6.0            # mean centroid distance to global axis (Å), heuristic
TARGET_ANGLE        = 60.0           # idealized pairwise angle (deg) for a “canonical” 4HB
MAX_HELIX_TILT_DEG  = 20.0           # tilt to bundle axis target (deg)
OUTPUT_CSV          = "foldability_metrics.csv"

def _check_mkdssp():
    if shutil.which("mkdssp") or shutil.which("dssp"):
        return
    raise RuntimeError(
        "mkdssp/dssp not found on PATH. Install DSSP first "
        "(e.g., conda install -c salilab dssp)."
    )

def _load_structure(pdb_path):
    parser = PDBParser(QUIET=True)
    return parser.get_structure(os.path.basename(pdb_path), pdb_path)

def _dssp_rows(pdb_path, chain_id=None):
    """
    Return ordered per-residue rows:
      (chain, resseq, icode, aa, ss, CAxyz, CA_bfactor)
    - ss uses DSSP code; ' ' -> '-' for convenience
    - CA_bfactor is used as pLDDT proxy if your PDB uses B-factor to store it (e.g., AlphaFold2).
    """
    _check_mkdssp()
    structure = _load_structure(pdb_path)
    model = structure[0]

    dssp = DSSP(model, pdb_path)

    rows = []
    for ch in model:
        cid = ch.id
        if (chain_id is not None) and (cid != chain_id):
            continue
        for res in ch:
            hetflag, resseq, icode = res.id
            if hetflag.strip():           # skip HETATM/water
                continue
            if 'CA' not in res:           # need CA for axes
                continue
            key = (cid, res.id)
            if key not in dssp:
                continue
            d = dssp[key]
            aa = d[1] if len(d) > 1 else 'X'
            ss = d[2] if len(d) > 2 else ' '
            ca = res['CA'].get_coord().astype(float)
            b  = float(res['CA'].get_bfactor())
            rows.append((cid, resseq, (icode.strip() or ' '), aa, (ss if ss != ' ' else '-'), ca, b))

    rows.sort(key=lambda t: (t[0], t[1], t[2]))
    return rows

def _contiguous_helices(ss_list, min_len=MIN_HELIX_LEN, helix_codes=HELIX_CODES):
    segs = []
    start = None
    for i, ss in enumerate(ss_list):
        if ss in helix_codes:
            if start is None:
                start = i
        else:
            if start is not None and (i - start) >= min_len:
                segs.append((start, i - 1))
            start = None
    if start is not None and (len(ss_list) - start) >= min_len:
        segs.append((start, len(ss_list) - 1))
    return segs

def _pca_axis(coords):
    X = coords - coords.mean(axis=0)
    _, _, Vt = np.linalg.svd(X, full_matrices=False)
    axis = Vt[0]
    return axis / np.linalg.norm(axis)

def _helix_centroid_axis(ca_coords, seg):
    pts = ca_coords[seg[0]:seg[1]+1]
    centroid = pts.mean(axis=0)
    axis = _pca_axis(pts)
    return centroid, axis

def _pairwise_angles_deg(axes):
    angs = []
    for i in range(len(axes)):
        for j in range(i+1, len(axes)):
            d = float(np.clip(abs(np.dot(axes[i], axes[j])), 0.0, 1.0))
            angs.append(np.degrees(np.arccos(d)))
    return np.array(angs, float) if angs else np.array([], float)

def _bundle_axis(centroids):
    C = np.vstack(centroids)
    return _pca_axis(C)

def _mean_radius_to_axis(centroids, axis):
    base = np.mean(np.vstack(centroids), axis=0)
    axis = axis / np.linalg.norm(axis)
    dists = []
    for c in centroids:
        v = c - base
        perp = v - np.dot(v, axis) * axis
        dists.append(np.linalg.norm(perp))
    return float(np.mean(dists)) if dists else np.nan

def _angles_to_bundle_axis(axes, bundle_axis):
    """Angles (deg) between each helix axis and the bundle axis."""
    angs = []
    ba = bundle_axis / np.linalg.norm(bundle_axis)
    for a in axes:
        a = a / np.linalg.norm(a)
        d = float(np.clip(abs(np.dot(a, ba)), 0.0, 1.0))
        angs.append(np.degrees(np.arccos(d)))
    return np.array(angs, float)

def analyze_pdb_4hb(pdb_path, chain_id=None):
    # Handle empty or malformed files early
    try:
        rows = _dssp_rows(pdb_path, chain_id=chain_id)
    except Exception as e:
        return {
            "file": os.path.basename(pdb_path),
            "n_res": np.nan,
            "n_helices": 0,
            "helix_lengths": [],
            "helix_frac": np.nan,
            "angles_mean": np.nan,
            "angles_sd": np.nan,
            "bundle_radius": np.nan,
            "tilt_mean": np.nan,
            "tilt_sd": np.nan,
            "length_cv": np.nan,
            "plddt_helix_mean": np.nan,
            "is_4hb": False,
            "error": str(e)
        }

    if not rows:
        return {
            "file": os.path.basename(pdb_path),
            "n_res": 0,
            "n_helices": 0,
            "helix_lengths": [],
            "helix_frac": np.nan,
            "angles_mean": np.nan,
            "angles_sd": np.nan,
            "bundle_radius": np.nan,
            "tilt_mean": np.nan,
            "tilt_sd": np.nan,
            "length_cv": np.nan,
            "plddt_helix_mean": np.nan,
            "is_4hb": False
        }

    ss = [r[4] for r in rows]
    ca = np.vstack([r[5] for r in rows])
    bf = np.array([r[6] for r in rows], float)  # potential pLDDT proxy

    segs = _contiguous_helices(ss, min_len=MIN_HELIX_LEN, helix_codes=HELIX_CODES)
    helix_lengths = [s[1] - s[0] + 1 for s in segs]
    n_helices = len(segs)
    helix_frac = float(sum(helix_lengths)) / max(1, len(ss))

    result = {
        "file": os.path.basename(pdb_path),
        "n_res": len(ss),
        "n_helices": n_helices,
        "helix_lengths": helix_lengths,
        "helix_frac": helix_frac,
        "tilt_mean": np.nan,
        "tilt_sd": np.nan,
        "length_cv": np.nan,
        "plddt_helix_mean": np.nan
    }

    if n_helices < REQ_N_HELICES:
        result.update({
            "angles_mean": np.nan,
            "angles_sd": np.nan,
            "bundle_radius": np.nan,
            "is_4hb": False
        })
        return result

    # Take 4 longest helices and compute geometry
    idx4 = np.argsort(helix_lengths)[::-1][:REQ_N_HELICES]
    segs4 = [segs[i] for i in idx4]

    centroids, axes = [], []
    for seg in segs4:
        c, a = _helix_centroid_axis(ca, seg)
        centroids.append(c); axes.append(a)

    angs = _pairwise_angles_deg(axes)
    angles_mean = float(np.mean(angs)) if angs.size else np.nan
    angles_sd   = float(np.std(angs))  if angs.size else np.nan

    g_axis = _bundle_axis(centroids)
    bundle_radius = _mean_radius_to_axis(centroids, g_axis)

    # Per-helix tilt vs bundle axis
    tilts = _angles_to_bundle_axis(axes, g_axis)
    tilt_mean = float(np.mean(tilts)) if tilts.size else np.nan
    tilt_sd   = float(np.std(tilts))  if tilts.size else np.nan

    # Length balance (CV) over the chosen 4 helices
    hl4 = np.array([s[1] - s[0] + 1 for s in segs4], float)
    length_cv = float(np.std(hl4) / max(np.mean(hl4), 1e-6)) if hl4.size else np.nan

    # Optional: mean pLDDT over helix residues (if B-factor encodes it)
    try:
        mask = np.zeros(len(ss), dtype=bool)
        for s in segs4:
            mask[s[0]:s[1]+1] = True
        # Heuristic: cap to [0,100]; ignore if wildly out of range
        plddt_vals = bf[mask]
        if plddt_vals.size and np.nanmax(plddt_vals) > 1.5:
            plddt_helix_mean = float(np.clip(np.nanmean(plddt_vals), 0.0, 100.0))
        else:
            plddt_helix_mean = np.nan
    except Exception:
        plddt_helix_mean = np.nan

    ok_mean = (ANGLE_RANGE_DEG[0] <= angles_mean <= ANGLE_RANGE_DEG[1]) if not np.isnan(angles_mean) else False
    ok_sd   = (angles_sd <= ANGLE_SD_MAX) if not np.isnan(angles_sd) else False
    ok_rad  = (bundle_radius <= BUNDLE_RADIUS_MAX_A) if not np.isnan(bundle_radius) else False

    result.update({
        "angles_mean": angles_mean,
        "angles_sd": angles_sd,
        "bundle_radius": bundle_radius,
        "tilt_mean": tilt_mean,
        "tilt_sd": tilt_sd,
        "length_cv": length_cv,
        "plddt_helix_mean": plddt_helix_mean,
        "is_4hb": bool(ok_mean and ok_sd and ok_rad)
    })
    return result

# -----------------------------
# Arm inference from filenames
# -----------------------------
def infer_arm_from_filename(path: str) -> str:
    fname = os.path.basename(path).lower()
    if   fname.startswith("cvae"):                       return "cVAE"
    elif fname.startswith("hybrid"):                     return "Hybrid"
    elif fname.startswith("interp") or fname.startswith("interpolation"): return "Interpolation"
    return "Unknown"

# -----------------------------
# Run batch analysis
# -----------------------------
files = sorted(glob.glob(INPUT_GLOB))
if not files:
    raise FileNotFoundError(f"No PDB files matched pattern: {INPUT_GLOB}")

rows = []
for p in files:
    metrics = analyze_pdb_4hb(p, chain_id=CHAIN_ID)
    metrics["arm"] = infer_arm_from_filename(p)
    rows.append(metrics)

df = pd.DataFrame(rows)
df.to_csv(OUTPUT_CSV, index=False)
print(f"Saved per-structure 4HB metrics: {OUTPUT_CSV}")
display(df.head(10))

# -----------------------------
# Scoring & Top-10 ranking
# -----------------------------
def _safe(x, default=np.nan):
    try:
        return float(x)
    except Exception:
        return default

def _soft_closeness_to_target(x, target, lo, hi):
    """
    1.0 at target; linearly decays to ~0.0 at the worst edge in [lo, hi].
    Values outside [lo,hi] are clamped to edges before measuring distance.
    """
    if np.isnan(x): return 0.0
    worst = max(abs(target - lo), abs(hi - target), 1e-6)
    clamped = min(max(x, lo), hi)
    return float(np.clip(1.0 - abs(clamped - target)/worst, 0.0, 1.0))

def _soft_better_lower(x, limit):
    """1.0 at 0; decays to 0.0 at 'limit' (linear hinge)."""
    if np.isnan(x): return 0.0
    return float(np.clip(1.0 - x/max(limit,1e-6), 0.0, 1.0))

def _soft_better_higher01(x):
    """Expect x in [0,1]; else clamp."""
    if np.isnan(x): return 0.0
    return float(np.clip(x, 0.0, 1.0))

def _soft_tilt(tilt_mean, max_tilt=MAX_HELIX_TILT_DEG):
    """1.0 when tilt_mean=0; ~0.0 when tilt_mean >= max_tilt."""
    if np.isnan(tilt_mean): return 0.0
    return float(np.clip(1.0 - tilt_mean/max(max_tilt,1e-6), 0.0, 1.0))

def _soft_length_balance(cv, cv_worst=0.5):
    """CV=0 -> 1.0; CV>=cv_worst -> ~0.0."""
    if np.isnan(cv): return 0.5
    return float(np.clip(1.0 - cv/max(cv_worst,1e-6), 0.0, 1.0))

def fourhb_quality_score_v2(row,
                            # geometry targets / limits
                            target_pair_angle=TARGET_ANGLE,
                            pair_angle_range=ANGLE_RANGE_DEG,
                            angle_sd_max=ANGLE_SD_MAX,
                            radius_max=BUNDLE_RADIUS_MAX_A,
                            max_tilt_deg=MAX_HELIX_TILT_DEG,
                            # --- Weights (geometry-forward) ---
                            w_pair_angle=0.30,
                            w_pair_sd=0.22,
                            w_radius=0.20,
                            w_tilt=0.15,
                            w_helix_frac=0.08,
                            w_length_balance=0.05,
                            # Optional pLDDT multiplier
                            use_plddt=True):
    """
    Composite 4HB score in [0,1]; higher is better.
    Emphasizes geometric regularity + compactness, with tilt & length balance.
    Optionally multiplies by (mean pLDDT over helix residues)/100 if available.
    """
    if not bool(row.get("is_4hb", False)):
        return np.nan

    # Pull metrics
    ang_mean = _safe(row.get("angles_mean"))
    ang_sd   = _safe(row.get("angles_sd"))
    radius   = _safe(row.get("bundle_radius"))
    hfrac    = _safe(row.get("helix_frac"))
    tilt_m   = _safe(row.get("tilt_mean"))
    length_cv= _safe(row.get("length_cv"))
    plddt_m  = _safe(row.get("plddt_helix_mean"))

    # Component scores (soft / continuous)
    s_pair   = _soft_closeness_to_target(ang_mean, target_pair_angle, *pair_angle_range)
    s_sd     = _soft_better_lower(ang_sd, angle_sd_max)
    s_rad    = _soft_better_lower(radius, radius_max)
    s_tilt   = _soft_tilt(tilt_m, max_tilt_deg)
    s_hfrac  = _soft_better_higher01(hfrac)
    s_lenbal = _soft_length_balance(length_cv, cv_worst=0.5)

    score = (
        w_pair_angle   * s_pair   +
        w_pair_sd      * s_sd     +
        w_radius       * s_rad    +
        w_tilt         * s_tilt   +
        w_helix_frac   * s_hfrac  +
        w_length_balance * s_lenbal
    )

    # Optional confidence scaling by pLDDT
    if use_plddt and not np.isnan(plddt_m) and 0.0 <= plddt_m <= 100.0:
        score *= (plddt_m / 100.0)

    return float(np.clip(score, 0.0, 1.0))

# Compute scores (NaN for non-4HB rows)
df["fourhb_score"] = df.apply(fourhb_quality_score_v2, axis=1)

# Top-20 among bona fide 4HBs
top20 = (
    df[df["is_4hb"] == True]
    .copy()
    .sort_values(["fourhb_score", "angles_sd", "bundle_radius"], ascending=[False, True, True])
    .head(20)
)

# Select helpful columns for display/saving
cols = [
    "file", "arm", "fourhb_score",
    "n_res", "n_helices", "helix_frac",
    "angles_mean", "angles_sd", "bundle_radius",
    "tilt_mean", "tilt_sd", "length_cv", "plddt_helix_mean"
]
top20_out = top20[cols].reset_index(drop=True)

print("\n=== Top-20 4HB configurations (higher score = better) ===")
display(top20_out)

# Save to CSV
top20_out.to_csv("top20_4hb.csv", index=False)
print("Saved Top-20 ranking to top20_4hb.csv")

# -----------------------------
# Summarize by arm 
# -----------------------------
if "arm" in df.columns:
    summ = (
        df.groupby("arm")["is_4hb"]
          .agg(['sum','count'])
          .rename(columns={'sum':'k_4hb','count':'n'})
          .sort_index()
    )
    summ["rate"] = summ["k_4hb"] / summ["n"].clip(lower=1)
    print("\n=== 4HB success by arm ===")
    display(summ)

    # Bar plot of 4HB rate by arm
    plt.figure(figsize=(5,4))
    plt.bar(summ.index.tolist(), summ["rate"].values)
    plt.ylim(0, 1)
    plt.ylabel("4-helix bundle success rate")
    plt.title("DSSP-based 4HB rate by arm")
    plt.grid(True, axis='y', linestyle=':', linewidth=0.6)
    plt.tight_layout()
    plt.show()
# -----------------------------
# Balanced resampling (robustness check)
# -----------------------------
import numpy as np
import pandas as pd

rng = np.random.default_rng(12345)

# Only keep rows with a valid arm label
df_bal = df[df["arm"].notna()].copy()
sizes = df_bal["arm"].value_counts()
n_min = sizes.min()  # balance to smallest arm size

R = 1000  # number of resamples
res_rate = []
res_score = []

# Compute:
#  - success rate per arm (mean of is_4hb)
#  - mean fourhb_score per arm, but only among true 4HBs (quality among successes)
for r in range(R):
    row = {}
    row_s = {}
    for arm, sub in df_bal.groupby("arm"):
        # sample without replacement to n_min
        pick = sub.sample(n=n_min, replace=False, random_state=int(rng.integers(1e9)))
        row[f"{arm}_rate"] = pick["is_4hb"].mean()

        ok = pick[pick["is_4hb"] == True]
        row_s[f"{arm}_score_mean"] = ok["fourhb_score"].mean() if len(ok) else np.nan

    res_rate.append(row)
    res_score.append(row_s)

res_rate = pd.DataFrame(res_rate)
res_score = pd.DataFrame(res_score)

# Aggregate: mean ± SD + 95% CI for rates and scores
def summarize_cols(df_in):
    out = []
    for c in df_in.columns:
        v = df_in[c].dropna().values
        if v.size == 0:
            out.append({"metric": c, "mean": np.nan, "sd": np.nan, "ci_lo": np.nan, "ci_hi": np.nan})
            continue
        m = float(np.mean(v))
        s = float(np.std(v, ddof=1))
        # normal approx CI on the resampling distribution
        z = 1.959963984540054
        half = z * s / np.sqrt(len(v))
        out.append({"metric": c, "mean": m, "sd": s, "ci_lo": m - half, "ci_hi": m + half})
    return pd.DataFrame(out)

rate_summary  = summarize_cols(res_rate)
score_summary = summarize_cols(res_score)

rate_summary.to_csv("resampling_rate_summary.csv", index=False)
score_summary.to_csv("resampling_score_summary.csv", index=False)

print("\n=== Balanced resampling (n = smallest arm) — 4HB success rate (mean ± SD; 95% CI over resamples) ===")
try:
    display(rate_summary)
except:
    print(rate_summary.to_string(index=False))

print("\n=== Balanced resampling (n = smallest arm) — fourhb_score among 4HBs (mean ± SD; 95% CI) ===")
try:
    display(score_summary)
except:
    print(score_summary.to_string(index=False))
```
